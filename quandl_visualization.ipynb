```python
import quandl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 給定seaborn的style
sns.set(style="whitegrid")
```

```python
# 抓出Apple和微軟2012~2017的股價
appl_df = quandl.get('WIKI/AAPL', start_date = '2012-01-01', end_date = '2017-10-31')
msft_df = quandl.get('WIKI/MSFT', start_date = '2012-01-01', end_date = '2017-10-31')
```

```python
# 預覽apple股價
appl_df.head()
```

```python
# 預覽微軟股價
msft_df.head()
```

```python
# 檢查資料
list(appl_df.index)
list(appl_df['Close'])
```

```python
# 作圖: x軸(日期)、y軸(收盤價)
plt.figure(figsize = (10, 8))
# 第一個子圖-Apple
plt.subplot(2, 1, 1)
plt.plot_date(x = appl_df.index, y = appl_df['Close'], fmt = 'r-', label = 'APPL')
plt.legend(loc = 'upper left')
# 第二個子圖-微軟
plt.subplot(2, 1, 2)
plt.plot_date(x = msft_df.index, y = msft_df['Close'], fmt = 'b-', label = 'MSFT')
plt.legend(loc = 'upper left')
```

```python
# 顯示
plt.show()
```
